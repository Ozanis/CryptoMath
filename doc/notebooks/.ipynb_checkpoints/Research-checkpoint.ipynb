```python
import ctypes
from os import getcwd
from Profiling.Plotting import *

modulo = 2**16 -1
test_name = "REDC"

fn.redc.argtypes = (ctypes.c_int, ctypes.c_int)

x1 = [0, 1, 2, 3, 2, 36, 5, 3, 32, 7, 2, 30, 17, 11]
y1 = [0, 1, 2, 2, 3, 9,  3, 5, 8,  2, 7,  5, 11, 17]
c1 = [(x1[i]*y1[i]) % modulo for i in range(len(y1))]

vals = load(x1, y1, c1)
plot(args = vals, title = test_name, fn = fn.redc)

x2 = [100, 1000, 1101, 555,  777, 777]
y2 = [101, 124,   555, 1101, 3,   8 ]
c2 = [(x2[i]*y2[i]) % modulo for i in range(len(x2))]

vals = load(x2, y2, c2)
plot(args = vals, title = test_name, fn = fn.redc)

```
