```python
import ctypes
import matplotlib.pyplot as plt 
from os import getcwd

test_lib = ctypes.CDLL(getcwd() + "/libMontgomery.so")
test_lib.gcd.argtypes = (ctypes.c_int, ctypes.c_int)

test_args1 = [0, 1, 2, 3, 2, 36, 5, 3, 32, 7, 2, 30, 17, 11]
test_args2 = [0, 1, 2, 2, 3, 9,  3, 5, 8,  2, 7,  5, 11, 17]

y = ["GCD({0}, {1})".format(test_args1[i], test_args2[i])  for i in range(len(test_args1))]
x = [test_lib.gcd(ctypes.c_int(test_args1[i]), ctypes.c_int(test_args2[i])) for i in range(len(y))]

plt.plot(x, y, color="c", linestyle="solid", marker="o", markerfacecolor='blue', markersize=12)
plt.title("GCD_test")
plt.grid()
plt.show()

#large numbers test

test_args1 = [100, 1000, 1101, 555,  777, 777]
test_args2 = [101, 124,   555, 1101, 3,   8 ]

y = ["GCD({0}, {1})".format(test_args1[i], test_args2[i])  for i in range(len(test_args1))]
x = [test_lib.gcd(ctypes.c_int(test_args1[i]), ctypes.c_int(test_args2[i])) for i in range(len(y))]
plt.plot(x, y, color="k", linestyle="solid", marker="o", markerfacecolor='green', markersize=12)
plt.title("GCD_test")
plt.grid()
plt.show()
```

```python
import ctypes
import matplotlib.pyplot as plt 
from os import getcwd

test_lib = ctypes.CDLL(getcwd() + "/libMontgomery.so")
test_lib.redc.argtypes = (ctypes.c_int, ctypes.c_int)

modulo = 65535

test_args1 = [0, 1, 2, 3, 2, 36, 5, 3, 32, 7, 2, 30, 17, 11]
test_args2 = [0, 1, 2, 2, 3, 9,  3, 5, 8,  2, 7,  5, 11, 17]

y = ["Redc({0}, {1})".format(test_args1[i], test_args2[i])  for i in range(len(test_args1))]
x = [test_lib.redc(ctypes.c_int(test_args1[i]), ctypes.c_int(test_args2[i])) for i in range(len(y))]
_x = [(test_args1[i]*test_args2[i]) % modulo for i in range(len(x))]

plt.figure()
#plt.subplot(132)
plt.plot(x, y, color="r", linestyle="solid", marker="o")
plt.title("redc_test")
plt.grid()
#plt.subplot(133)
plt.plot(_x, y, color="c", linestyle="solid", marker='o')
plt.suptitle('Comparison')
plt.show()


test_args1 = [100, 1000, 1101, 555,  777, 777]
test_args2 = [101, 124,   555, 1101, 3,   8 ]

y = ["Redc({0}, {1})".format(test_args1[i], test_args2[i])  for i in range(len(test_args1))]
x = [test_lib.redc(ctypes.c_int(test_args1[i]), ctypes.c_int(test_args2[i])) for i in range(len(y))]
_x = [(test_args1[i]*test_args2[i]) % modulo for i in range(len(x))]

plt.figure()
#plt.subplot(132)
plt.plot(x, y, color="r", linestyle="solid", marker="o")
plt.title("redc_test")
plt.grid()
#plt.subplot(133)
plt.plot(_x, y, color="c", linestyle="solid", marker='o')
plt.suptitle('Comparison')
plt.show()        
```
